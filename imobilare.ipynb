```python
# Data from https://data.europa.eu/data/datasets/2db150a8-2995-4dff-bb35-c1dd1f8dbd76?locale=en

import seaborn as sns
import pandas as pd
sns.set_theme(style="whitegrid")

df = pd.read_csv('~/Documents/Presentations/seaborn_tutorial/datasets/tranzactii-imobiliare-sectorul-2-2020.csv', sep=';')
df = df[df["Suprafata utila"] < 150]
df = df[df["tipProprietate"] ==  'apartament']
#df = df[df['Zona'] == colentina]
df
```

```python
g = sns.relplot(data=df, x="Suprafata utila", y="Pret tranz")
```

```python
g = sns.relplot(data=df, x="Suprafata utila", y="Pret tranz", hue="Zona")
```

```python
g = sns.boxplot(data=df, x="Pret tranz", y="Zona")
```

```python
grid = sns.FacetGrid(df, col="Tip strada", col_wrap=4)
grid.map(sns.scatterplot, "Suprafata utila", "Pret tranz")
```
